### Term -> DocID

```python
dic = []
for f in docs:
    # Traverse the docs
    doc = open(pth+f,'r',encoding='utf8')
    lines = doc.readlines()
    doc_name = f.split()
    doc_name = doc_name[-1][:-4] 
    # 去掉.txt和外部引号
    for line in lines:
        words = line.split()
        for w in words:
            dic.append((w.lower(),doc_name))
            # gai terms
```

```python
dic = sorted(dic, key=lambda s: s[0])
term = dic[0][0] # 处理的词项
docid = [dic[0][1]] # DocID
```

```python
ans = []
for i in dic[1:]:
    if term == i[0]:
        if docid[-1] != i[1]:
            docid.append(i[1])
            # calc times
    else:
        docid = sorted(docid,key=lambda s:int(s))
        ans.append((term,len(docid),docid))
        term = i[0]
        # use array
        docid = [i[1]]
```

```python
ans = sorted(ans,
             key=lambda s:int(s[1]),
             reverse=False)

# 升序降序都可以
descending_ans = sorted(ans,
             key=lambda s:int(s[1]),
             reverse=True)
```
